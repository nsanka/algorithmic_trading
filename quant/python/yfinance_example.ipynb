```python
import yfinance as yf
import talib as ta
tesla_data = yf.download('TSLA', '2024-01-01', '2024-10-13')
tesla_data['SMA'] = ta.SMA(tesla_data['Adj Close'], 14)
tesla_data.SMA.plot();
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

plt.style.use('seaborn-v0_8-darkgrid')
tesla_data['Close'].plot(figsize=(10, 7), color='green')
plt.title('Closing Price', fontsize=14)
plt.xlabel('Year-Month', fontsize=12)
plt.ylabel('Price ($)', fontsize=12)
plt.show()
```
